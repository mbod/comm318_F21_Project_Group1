```python
# ONLY IMPORT MODULES YOU ARE GOING TO USE IN NOTEBOOK

import pandas as pd
import matplotlib.pyplot as plt
import math
import seaborn as sns
import numpy as np
```

```python
health_df = pd.read_csv("../data/health_clean.csv")
```

## Question One

```python
health_df.plot(kind='scatter', x = 'poor_fair_health', y = 'life_exp')
plt.xlabel("Overal Health")
plt.ylabel("Life Expectancy")
plt.title("Life Expectancy by Overall Health")
plt.show()
health_df.plot(kind='scatter', x = 'poor_fair_health', y = 'poor_physical_days')
plt.xlabel("Overal Health")
plt.ylabel("Physical Health")
plt.title("Physical Health by Overall Health")
plt.show()
health_df.plot(kind='scatter', x = 'poor_fair_health', y = 'poor_mental_days')
plt.xlabel("Overal Health")
plt.ylabel("Mental Health")
plt.title("Mental Health by Overall Health")
plt.show()
```

```python
health_df[['poor_fair_health', 'life_exp', 'poor_physical_days','poor_mental_days']].corr()
```

### Answer and Interpretation/Observation

The outcomes are as expected. Overall health has a strong positive correlation to all other measures of health. That is to say as overall health increases, physical health, mental health, and life expectancy all increase as well. From the correlation table, physical and mental health share the highest correlation.

```python
health_df_state = health_df.groupby('state_abbreviation')[['poor_fair_health', 'poor_physical_days', 'poor_mental_days', 'life_exp']].mean()
```

## Question 2

```python
plt.rcParams["figure.figsize"]=12,4

health_df_state.plot(kind='bar', y = 'life_exp')

plt.xlabel("")
plt.ylabel("Life Expectancy")
plt.title("Life Expectancy by State")
plt.xticks(rotation = 90)
plt.legend().remove()
plt.show()
health_df_state.plot(kind='bar', y = 'poor_fair_health')
plt.xlabel("")
plt.ylabel("Health")
plt.title("Overal Health by State")
plt.xticks(rotation = 90)
plt.legend().remove()
plt.show()
health_df_state.plot(kind='bar', y = 'poor_physical_days')
plt.xlabel("")
plt.ylabel("Pyhsical Health")
plt.title("Pyhsical Health by State")
plt.xticks(rotation = 90)
plt.legend().remove()
plt.show()
health_df_state.plot(kind='bar', y = 'poor_mental_days')
plt.xlabel("")
plt.ylabel("Mental Health")
plt.title("Mental Health by State")
plt.xticks(rotation = 90)
plt.legend().remove()
plt.show()
```

### Answer and Interpretation/Observation

No state particually stands out in regards to life expectancy. However, West Virginia has the poorest average mental and physical health (it is not surprising that the same state has the poorest mental and physical health given the high correlation between the too seen in question 1). Additionally, Mississippi has the poorest overall health.
